```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([22,34,55,27,88])
y = np.array([21,64,85,29,58])
z = np.array([25,94,65,27,68])
```

```python
#line plot
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x , y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 10);
plt.grid()
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(y , z, marker = '>', mfc = 'red', color = 'black', ls = '--', ms = 10);
plt.grid()
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(y , z, marker = '>', mfc = 'red', color = 'black', ls = '--', ms = 10, label = '2023');
plt.plot(x , y, marker = 'o', mfc = 'blue', color = 'pink', ls = '--', ms = 10, label = '2024');
plt.grid()
plt.legend()
```

```python
#subplot

plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x , y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 10);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(y , z, marker = '>', mfc = 'red', color = 'black', ls = '--', ms = 10);
plt.grid()
```
